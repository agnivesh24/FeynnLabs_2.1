```python
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
```

```python
# Assuming MD_x is already defined as in the previous code:
# Example data loading and preparation (if not already defined)
# mcdonalds = pd.read_csv('mcdonalds.csv')
# MD_x = mcdonalds.iloc[:, 0:11]
# MD_x = (MD_x == "Yes").astype(int)

# Set the random seed for reproducibility
np.random.seed(1234)

# Bootstrapping settings
n_clusters_range = range(2, 9)  # Corresponding to 2:8 in R
n_rep = 10  # Number of replications
n_boot = 100  # Number of bootstraps
results = {}

# Perform bootstrapped clustering and calculate adjusted Rand index
for n_clusters in n_clusters_range:
    scores = []
    for _ in range(n_rep):
        # Sample with replacement (bootstrap sampling)
        X_boot = resample(MD_x, replace=True, n_samples=len(MD_x))
        # Fit KMeans
        kmeans = KMeans(n_clusters=n_clusters, random_state=np.random.randint(0, 10000))
        kmeans.fit(X_boot)
        labels_boot = kmeans.labels_

        # Fit KMeans on the original data
        kmeans_orig = KMeans(n_clusters=n_clusters, random_state=np.random.randint(0, 10000))
        kmeans_orig.fit(MD_x)
        labels_orig = kmeans_orig.labels_

        # Calculate adjusted Rand index between original and bootstrapped labels
        score = adjusted_rand_score(labels_orig, labels_boot)
        scores.append(score)

    # Store the mean of the adjusted Rand index for each number of clusters
    results[n_clusters] = np.mean(scores)

# Plot the results
plt.figure(figsize=(8, 6))
plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand Index')
plt.title('Bootstrapped Clustering Stability')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import entropy

np.random.seed(1234)
k_values = range(2, 9)
MD_m28 = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=1234)
    model.fit(MD_x.values)
    iter_val = model.n_iter_
    converged = True
    k_val = k
    k0_val = k
    log_likelihood = -model.inertia_
    n_samples, _ = MD_x.shape
    aic = -2 * log_likelihood + 2 * k
    bic = -2 * log_likelihood + np.log(n_samples) * k
    labels = model.labels_
    counts = np.bincount(labels)
    probs = counts / float(counts.sum())
    class_entropy = entropy(probs)
    icl = bic - class_entropy

    MD_m28.append((iter_val, converged, k_val, k0_val, log_likelihood, aic, bic, icl))
MD_m28 = pd.DataFrame(MD_m28, columns=['iter', 'converged', 'k', 'k0', 'logLik', 'AIC', 'BIC', 'ICL'])

print(MD_m28)
```

```python
num_segments = MD_m28["k"]
AIC_values = MD_m28["AIC"]
BIC_values = MD_m28["BIC"]
ICL_values = MD_m28["ICL"]

plt.plot(num_segments, AIC_values, marker='o', label='AIC')
plt.plot(num_segments, BIC_values, marker='o', label='BIC')
plt.plot(num_segments, ICL_values, marker='o', label='ICL')

plt.xlabel('Number of Segments')
plt.ylabel('Value of Information Criteria')
plt.title('Information Criteria (AIC, BIC, ICL)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

def labelling(x):
    mcdonalds[x] = LabelEncoder().fit_transform(mcdonalds[x])
    return mcdonalds
cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    labelling(i)

df_eleven = mcdonalds.loc[:, cat]

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df_eleven)
mcdonalds['cluster_num'] = kmeans.labels_

crosstab = pd.crosstab(mcdonalds['cluster_num'], mcdonalds['Like'])

like_order = ['I HATE IT!-5', '-4', '-3', '-2', '-1', '0', '1', '2', '3', '4', 'I LOVE IT!+5']
crosstab = crosstab.reindex(columns=like_order, fill_value=0)

plt.rcParams['figure.figsize'] = (10, 7)
mosaic(crosstab.stack(), title='Cluster vs Like Mosaic Plot')
plt.show()
```
